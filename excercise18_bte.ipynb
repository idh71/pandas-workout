```python
import pandas as pd
import numpy as np
```

```python
# to ignore certian columns, you still include them in the names parameter but leave them out of usecols
df = pd.read_csv('/content/linux-etc-passwd.txt', sep=':',
                 comment='#',
                 header=None,
                 names=['username', 'password', 'userid', 'groupid',
                        'name', 'homedir', 'shell'],
                 usecols=['username', 'userid',
                        'name', 'homedir', 'shell'])
```

```python
# show the nonspecial usernames (those with userid >1000)
df.loc[(df['userid'] > 1000), ['userid', 'username']]
```

```python
df.shell.unique()
```

```python
# The `unique` method does this for us, returning a NumPy array -- but Pandas offers its own
# drop_duplicates method, which keeps us with a series
df['shell'].drop_duplicates()
```
