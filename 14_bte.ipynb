```python
import pandas as pd

```

```python
s = pd.read_csv('./pandas-workout-data/nyc-temps.txt').squeeze('columns')

df = pd.DataFrame({'temp': s,
                  'hour': [0, 3, 6, 9, 12, 15, 18, 21] * 91
                  })
```

```python
# set all the odd temps to the mean

df.loc[df['temp'] % 2 != 0, 'temp'] = df['temp'].mean()
```

```python
# just resetting the data
s = pd.read_csv('./pandas-workout-data/nyc-temps.txt').squeeze('columns')

df = pd.DataFrame({'temp': s,
                  'hour': [0, 3, 6, 9, 12, 15, 18, 21] * 91
                  })
```

```python
# set the even temperatures at 9 and 18 to 3
df.loc[(df['temp'] % 2 == 0) & (df['hour'] == 9) | (df['hour'] == 18 ), 'temp'] = 3
```
