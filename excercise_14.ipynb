```python
import numpy as np
import pandas as pd
```

```python
s = pd.read_csv("/content/nyc-temps.txt").squeeze()
```

```python
df  = pd.DataFrame({"temp": s,
                   "hour": [0, 3, 6, 9, 12, 15, 18, 21] * 91})
```

```python
# get the original mean and median
real_mean_temp = df.temp.mean()
real_mean_temp
```

```python
real_median_temp = df.temp.median()
real_median_temp
```
